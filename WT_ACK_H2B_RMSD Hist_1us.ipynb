```python
import MDAnalysis as mda
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import MDAnalysis.analysis.rms

#Select the path to .prmtop and .dcd files to analyze RMSD 
top='/media/rutika/My Book/analysis_use/dry_1kx5_015M_OPC.prmtop'
traj='/media/rutika/My Book/analysis_use/dry_wild_015_unwrap_1us_align.dcd'
u1 = mda.Universe(top,traj)

top1='/media/rutika/My Book/analysis_use/dry_ack_015M.prmtop'
traj1='/media/rutika/My Book/analysis_use/dry_ack_015_unwrap_1us_align.dcd'
u2 = mda.Universe(top1,traj1)
```

```python
#Select the resid of the residues of interest to analyze for RMSD for WT
import MDAnalysis.analysis.rms
bb = u1.select_atoms("resid 660-692 and name CA")
u1.trajectory[0]
A = bb.positions.copy()
RMSD=[]
for ts in u1.trajectory:
    B = bb.positions.copy()
    RMSD.append((ts.frame, MDAnalysis.analysis.rms.rmsd(A, B, center=True)))
print(RMSD)
```

```python
# import numpy library
import numpy
 
# create an array
a = RMSD
 
# save array into csv file
numpy.savetxt("RMSD_wt_015_1.csv", a,
              delimiter = ",")
rmsd_values = pd.read_csv('RMSD_wt_015_1.csv', usecols = [1])
rmsd_values
```

```python
#Select the resid of the residues of interest to analyze for RMSD for another ACK system
import MDAnalysis.analysis.rms
bb = u2.select_atoms("resid 366-398 and name CA")
u2.trajectory[0]
G = bb.positions.copy()
RMSD2=[]
for ts in u2.trajectory:
    H = bb.positions.copy()
    RMSD2.append((ts.frame, MDAnalysis.analysis.rms.rmsd(G, H, center=True)))
print(RMSD2)
```

```python
# import numpy library
import numpy
 
# create an array
b = RMSD2
 
# save array into csv file
numpy.savetxt("RMSD_ack_015_1.csv", b,
              delimiter = ",")
rmsd_values2 = pd.read_csv('RMSD_ack_015_1.csv', usecols = [1])
rmsd_values2
```

```python
#https://seaborn.pydata.org/generated/seaborn.distplot.html
import seaborn as sns

plt.subplots(figsize=(10, 7))
sns.distplot(rmsd_values, hist = False, kde = True, color='blue', label='WT_H2B1',
                 kde_kws = {'shade': True, 'linewidth': 3})
sns.distplot(rmsd_values2, hist = False, kde = True, color='orange', label='ACK_H2B1',
                 kde_kws = {'shade': True, 'linewidth': 3})
plt.xlabel('RMSD ($\AA$)', fontsize = 18.0, fontweight="bold")
plt.ylabel('Probability Density', fontsize = 18.0, fontweight="bold")
plt.tick_params(labelsize=15, pad=1)
plt.rcParams["font.weight"] = "bold"
plt.rcParams["axes.labelweight"] = "bold"
plt.legend(loc='upper right', fontsize='large', title='0.15M', title_fontsize='large')
plt.savefig("RMSD_histH2B1_1us_0.15m.svg", format="svg",transparent=True, dpi = 600)
plt.show()
```
