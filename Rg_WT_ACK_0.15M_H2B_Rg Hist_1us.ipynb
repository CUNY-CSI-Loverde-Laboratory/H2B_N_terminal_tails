```python
import MDAnalysis as mda
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#top='/home/rutika/nucleosome/production/wildtype/sasa/sasa_1us/dry_1kx5_015M_OPC.prmtop'
#traj='/home/rutika/nucleosome/production/wildtype/sasa/sasa_1us/dry_wild_015_unwrap_1us_align.dcd'

top='/media/rutika/My Book/analysis_use/1kx5_015M_OPC.prmtop'
traj='/media/rutika/My Book/analysis_use/1kx5_wild_015_unwrap_1us.dcd'
u1 = mda.Universe(top,traj)

#top1='/home/rutika/nucleosome/production/ack/sasa/1us_sasa/dry_ack_015M.prmtop'
#traj1='/home/rutika/nucleosome/production/ack/sasa/1us_sasa/dry_ack_015_unwrap_1us_align.dcd'

top1='/media/rutika/My Book/analysis_use/1kx5_ack.prmtop'
traj1='/media/rutika/My Book/analysis_use/1kx5_ack_015_unwrap_1us.dcd'
u2 = mda.Universe(top1,traj1)
```

```python
#WT H2B Tail 1
A = []
B =u1.select_atoms("resid 660-692 and name CA")
for ts in u1.trajectory:
    A.append((ts.frame, B.radius_of_gyration()))

print(A)
```

```python
# import numpy library
import numpy
 
# create an array
a = A
 
# save array into csv file
numpy.savetxt("Rg_wt_015_1.csv", a,
              delimiter = ",")
rg_values = pd.read_csv('Rg_wt_015_1.csv', usecols = [1])
rg_values
```

```python
#ACK H2B Tail 1
D = []
E =u2.select_atoms('resid 366-398 and name CA')
for ts in u2.trajectory:
    D.append((ts.frame, E.radius_of_gyration()))

print(D)
```

```python
# import numpy library
import numpy
 
# create an array
b = D
 
# save array into csv file
numpy.savetxt("Rg_ack_015_1.csv", b,
              delimiter = ",")
rg_values1 = pd.read_csv('Rg_ack_015_1.csv', usecols = [1])
rg_values1
```

```python
#https://seaborn.pydata.org/generated/seaborn.distplot.html
import seaborn as sns

plt.subplots(figsize=(10, 7))
sns.distplot(rg_values, hist = False, kde = True, color='blue', label='WT_H2B1',
                 kde_kws = {'shade': True, 'linewidth': 3})
sns.distplot(rg_values1, hist = False, kde = True, color='orange', label='ACK_H2B1',
                 kde_kws = {'shade': True, 'linewidth': 3})
plt.xlabel('Rg ($\AA$)', fontsize = 18.0, fontweight="bold")
plt.ylabel('Probability Density', fontsize = 18.0, fontweight="bold")
plt.tick_params(labelsize=15, pad=1)
plt.rcParams["font.weight"] = "bold"
plt.rcParams["axes.labelweight"] = "bold"
plt.legend(loc='upper right', fontsize='large', title='0.15M', title_fontsize='large')
plt.savefig("Rg_histH2B1_1us_0.15m.svg", format="svg",transparent=True, dpi = 600)
plt.show()
```
