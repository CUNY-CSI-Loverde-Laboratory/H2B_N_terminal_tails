```python
import MDAnalysis as mda
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#select the path to the .prmtop and .dcd files to analyze Rg 

top='/media/rutika/My Book/nucleosome_traj/wt_015m/dry_1kx5_015M_OPC.prmtop'
traj='/media/rutika/My Book/nucleosome_traj/wt_015m/dry_1kx5_wild_015_unwrap_1us.xtc'
u1 = mda.Universe(top,traj)


top1='/media/rutika/My Book/nucleosome_traj/wt_24m/dry_1kx5_2M_OPC_trial3.prmtop'
traj1='/media/rutika/My Book/nucleosome_traj/wt_24m/1kx5_wt_2_4_1us.xtc'
u2 = mda.Universe(top1,traj1)
```

```python
#Select the residues resid to analyze Rg 
A = []
B =u1.select_atoms("resid 660-692 and name CA")
for ts in u1.trajectory:
    A.append((ts.frame, B.radius_of_gyration()))

print(A)
```

```python
# import numpy library
import numpy
 
# create an array
a = A
 
# save array into csv file
numpy.savetxt("Rg_wt_h2b1_015_1.csv", a,
              delimiter = ",")
rg_values = pd.read_csv('Rg_wt_h2b1_015_1.csv', usecols = [1])
rg_values
```

```python
#Select the residues resid to analyze Rg 
#WT H4 Tail 1 2.4 M
D = []
E =u2.select_atoms('resid 660-692 and name CA')
for ts in u2.trajectory:
    D.append((ts.frame, E.radius_of_gyration()))

print(D)
```

```python
# import numpy library
import numpy
 
# create an array
b = D
 
# save array into csv file
numpy.savetxt("Rg_wt_24_H2b_1.csv", b,
              delimiter = ",")
rg_values1 = pd.read_csv('Rg_wt_24_H2b_1.csv', usecols = [1])
rg_values1
```

```python
import seaborn as sns

plt.subplots(figsize=(10, 7))
sns.distplot(rg_values, hist = False, kde = True, color='blue', label='WT_H2B_015',
                 kde_kws = {'shade': True, 'linewidth': 3})
sns.distplot(rg_values1, hist = False, kde = True, color='green', label='WT_H2B_24',
                 kde_kws = {'shade': True, 'linewidth': 3})
plt.xlabel('Rg ($\AA$)', fontsize = 18.0, fontweight="bold")
plt.ylabel('Probability Density', fontsize = 18.0, fontweight="bold")
plt.tick_params(labelsize=15, pad=1)
plt.rcParams["font.weight"] = "bold"
plt.rcParams["axes.labelweight"] = "bold"
plt.legend(loc='upper right', fontsize='large', title_fontsize='large')
plt.savefig("Rg_histH2b_tail1_1us.svg", format="svg",transparent=True, dpi = 600)
plt.show()
```
